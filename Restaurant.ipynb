## Cargar modulos

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
```

### Cargando Datos

```python
restaurant = pd.read_csv('Restaurante.csv')
restaurant.head(n=10)
```

```python
restaurant.columns
```

```python
rest= restaurant.drop( columns=["n"], axis=0)
```

## Insepeccionar datos

```python
rest.describe()
```

```python
rest.shape
```

## Evaluacion de datos faltantes

```python
rest.isnull().sum(axis=0)
```

## Exploracion de analisis de dato

```python
pd.crosstab(rest["restaurant"], columns= "count")
```

```python
pd.crosstab(rest["item"], columns= "count")
```

```python
pd.crosstab(rest["salad"], columns= "count")
```

## Creación de subconjuntos por palabras contenidas

```python
rest_anburguesa = rest[rest["item"].str.contains("Burger|burger")]
rest_anburguesa.head()
```

```python
rest_burrito = rest[rest["item"].str.contains("Burrito")]
rest_burrito.head()
```

```python
rest_sandwich = rest[rest["item"].str.contains("Sandwich")]
rest_sandwich.head()
```

## Ingenierria de Funciones

```python
rest['Mi_grupo'] = rest['item'].apply(lambda x: 1 if 'urger' in x else 2 if 'Sandwich' in x else 0)
rest.head()
```

## Variables numericos

## Boxplot con Seaborn

```python
plt.figure(figsize=(8,5))
sb.boxplot(x='restaurant', y='calories', data=rest, palette="cool")
plt.title('Contenido de calorias por Restaurante')
```
